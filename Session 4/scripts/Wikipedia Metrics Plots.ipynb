```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

```python
fn = "../output/wikipedia_metrics.csv"
df = pd.read_csv(fn)
```

```python
df.head()
```

```python
x = df["users"]
y = df["edits"]
```

```python
plt.plot(x,y, color = 'red', marker = 'o', alpha=0.5, linestyle='')
```

```python
plt.plot(x,y, c = 'k', marker = '.', ls='')
```

```python
plt.plot(x,y, 'k.')
```

```python
plt.plot(x,y, 'k.')
plt.xlabel("No. of Contributing Users")
plt.ylabel("No. of Edits")
plt.minorticks_on()
plt.title("Forbes 100 Celebrities' Wikipedia Pages")
```

```python
mean_ratio = np.mean(y/x)
mean_ratio
```

```python
xx = np.linspace(800,8400,1000)
yy = mean_ratio * xx
```

```python
plt.plot(x,y,'r', marker = "o", alpha=0.3)
plt.plot(xx,yy,'b--',label="mean ratio = %.2f" % mean_ratio)

plt.xlabel("No. of Contributing Users")
plt.ylabel("No. of Edits")
plt.minorticks_on()
plt.title("Forbes 100 Celebrities' Wikipedia Pages")
plt.legend(loc="upper left", numpoints = 1)
```

```python
x_mean = np.mean(x)
x_mean
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x-x_mean,y)
```

```python
slope, intercept
```

```python
r_value, p_value, std_err
```

```python
xx_fit = xx - x_mean
yy_fit = slope*xx_fit + intercept
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,'ko', alpha = 0.3)
#plt.plot(xx,yy,'b--',label="mean ratio = %.2f" % mean_ratio)
plt.plot(xx,yy_fit,'r-',label="linreg slope = %.2f" % slope)

plt.xlabel("No. of Contributing Users")
plt.ylabel("No. of Edits")
plt.minorticks_on()
plt.title("Forbes 100 Celebrities' Wikipedia Pages")
plt.legend(loc="upper left", numpoints = 1)
plt.ylim((0,27000))

plt.savefig("../fig/fig_wikipedia_users_vs_no_of_edits.png")
```
